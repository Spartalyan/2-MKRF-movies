```python
movies['gos_fin_perc'] = (movies['refundable_support'] + movies['nonrefundable_support']) / movies['budget']
gosfin_perc = movies['gos_fin_perc'].median().round(2)
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø–æ–¥–¥–µ—Ä–∂–∫–∏ —Ñ–∏–ª—å–º–æ–≤ –≥–æ—Å—É–¥–∞—Ä—Å—Ç–≤–æ–º', gosfin_perc)
```

```python
movies.loc[incorrect_budget.index, 'budget'] = \
(movies.loc[incorrect_budget.index, 'refundable_support'] + movies.loc[incorrect_budget.index, 'nonrefundable_support'])\
/ gosfin_perc
movies.loc[incorrect_budget.index]
```

```python
#–ø—Ä–æ–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤—ã–±–∏—Ä–∞—Ç—å –ø–µ—Ä–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ
def main(name):
    name = name.split(',')
    return name[0]

#—Å—Ç–æ–ª–±—Ü—ã –æ—Ç–∫—É–¥–∞ –±—É–¥–µ–º –±—Ä–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –¥–ª—è –Ω–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
change_col = ['genres', 'director']

for col in change_col:
    movies['main_' + col] = movies[col].apply(main) 

    
def find_first (name):
    name = name.split(',')
    return name[0]
 
cols_2 = ['genres', 'director']
 
for col in cols_2:
    movies['new'] = movies[col].apply(find_first)
   
movies['new_' + col] = movies[col].apply(find_first) 
```
